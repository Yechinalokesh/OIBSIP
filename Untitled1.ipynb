```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
clf=DecisionTreeClassifier(random_state=42)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred) # Corrected function name to accuracy_score
print(f"Accuracy: {accuracy*100:.2f}%")
```

```python
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=list(iris.target_names), filled=True)
plt.title("Decision Tree for Iris Dataset", color='red',size=42)
plt.show()

```
